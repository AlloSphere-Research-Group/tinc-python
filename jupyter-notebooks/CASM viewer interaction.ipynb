```python
graph_db = tclient.get_disk_buffer("graph1")
```

```python
%pylab inline
```

```python
figname = graph_db.get_filename_for_writing()
plot(arange(1000))
title('Hello **********')
savefig(figname)
graph_db.done_writing_file(figname)
```

```python
import random
import matplotlib.pyplot as plt

def change_graph(value):
    figname = graph_db.get_filename_for_writing()
    f = plt.figure()
    plt.plot(arange(300))
    plt.title(f'Temperature PPPPP {value}')
    plt.savefig(figname)
    plt.close()
    f.clf()
    del f
    graph_db.done_writing_file(figname)

tclient.get_parameter("T").register_callback(change_graph)
```
